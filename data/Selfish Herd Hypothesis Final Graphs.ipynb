# Density dependent survival

$P_{\rm{capture}} = \frac{1}{A_{\rm{density}}}$

$A_{\rm{density}}$ is the number of prey within 30 units of the target prey, including the target prey itself.

```python
from pandas import *
import glob, numpy as np

figure(figsize=(7, 3.75))
ax = axes()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
xlabel("Number of Prey", fontsize="12")
ylabel("Predator Attack Efficiency", fontsize="12")
xticks(range(51)[::5])
yticks(arange(0, 1.01, 0.1))
xlim(0, 51)
ylim(0, 1)

A_NV = range(1, 51)

AttackEfficiency = []

for Density in A_NV:
    AttackEfficiency.append(1.0 / float(Density))

plot(A_NV, AttackEfficiency, "k-", label="A$_{nearby}$", lw=2)

#legend(prop={'size':12})

savefig("selfish-herd/final-graphs/density-dependent-survival.pdf")
```

# Artificial selection

```python
from pandas import *
import glob
import numpy as np

dataLists = {}

# read data
for folder in glob.glob("selfish-herd/swarm-sd30-selfish-herd-artificial-selection-*-attack-kd4"):
    
    dataLists[folder.split("/")[1]] = []
    
    for datafile in glob.glob(folder + "/*.csv"):
        
        replicateDF = read_csv(datafile)
        
        # sqrt the avg shortest distance because the simulation works in distance squared
        #replicateDF["avg_shortest_dist"] = replicateDF["avg_shortest_dist"].apply(np.sqrt)
        
        dataLists[folder.split("/")[1]].append(replicateDF)

# calculate stats for data
meanDFs = {}
stderrDFs = {}

for key in dataLists.keys():
    
    keyDF = (concat(dataLists[key], axis=1, keys=range(len(dataLists[key])))
            .swaplevel(0, 1, axis=1)
            .sortlevel(axis=1)
            .groupby(level=0, axis=1))
    
    meanDFs[key] = keyDF.mean()
    stderrDFs[key] = keyDF.std().div(sqrt(len(dataLists[key]))).mul(2.0)
    
    keyDF = None

# plot data
for column in meanDFs[key].columns:
    
    #keyList = sorted(meanDFs.keys())
    keyList = ["swarm-sd30-selfish-herd-artificial-selection-rand-attack-kd4", "swarm-sd30-selfish-herd-artificial-selection-randwalk-attack-kd4", "swarm-sd30-selfish-herd-artificial-selection-outside-attack-kd4"]
    
    if (column == "swarm_density_count"):
        
        figure(figsize=(7, 3.75), frameon=False)
        
        if column == "swarm_density_count":
            ylabel("Mean Nearby Prey", fontsize="12")
            ylim(-1, 40)

        xlabel("Generation", fontsize="12")
        
        ax = axes()
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.get_xaxis().tick_bottom()
        ax.get_yaxis().tick_left()
        
        xticks(np.arange(0, 1201, 100), fontsize="12")
        yticks(fontsize="12")
        xlim(-20, 1220)
        
        for key in keyList:
            
            keyLabel = None
            
            if key == "swarm-sd30-selfish-herd-artificial-selection-rand-attack-kd4":
                keyLabel = "Random Attack"
                barFormat = "^-"
                barColor = "0.7"
            elif key == "swarm-sd30-selfish-herd-artificial-selection-randwalk-attack-kd4":
                keyLabel = "Random Walk Attack"
                barFormat = "o--"
                barColor = "0.35"
            elif key == "swarm-sd30-selfish-herd-artificial-selection-outside-attack-kd4":
                keyLabel = "Outside Attack"
                barFormat = "s:"
                barColor = "0.0"
            
            errorbar(x=meanDFs[key]["generation"][::100], y=meanDFs[key][column][::100], yerr=stderrDFs[key][column][::100], label=keyLabel, fmt=barFormat, color=barColor, ms=6)
            
        legend(loc=2, prop={'size':12})
        savefig("selfish-herd/final-graphs/" + column.replace("_", "-") + "-artificial-selection.pdf")
```
